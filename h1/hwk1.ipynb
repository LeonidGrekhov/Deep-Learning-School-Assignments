```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import tensorflow as tf

print(tf.__version__)
```

```python
data= pd.read_csv("diabetes.csv", header=None) #file import
```

```python
data
```

```python
sns.set(rc={'figure.figsize':(12,12)})#quick matrix to see data in a more useful way
corr_matrix = data.corr().round(2)
sns.heatmap(data=corr_matrix, annot=True)
```

```python
data.shape
```

```python
from sklearn import preprocessing

x = data.iloc[:, data.columns != 8]
y = data.iloc[:, [8]]
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size =0.2, random_state =1)#spliting data
```

```python
linear_model = tf.keras.Sequential([tf.keras.layers.Dense(1)])
```

```python
# Get Loss from library
example_loss = tf.keras.losses.MeanSquaredError()
# Compile the model with an optimizer
linear_model.compile(optimizer='SGD',
                     loss=example_loss,
                     metrics=['accuracy'])
```

```python
linear_predict = linear_model.predict(X_test)
```

```python
# Create a Deep Model
sample_model = tf.keras.Sequential([
                                    tf.keras.layers.Dense(16,activation='sigmoid',input_shape=[8,]),
                                    tf.keras.layers.Dense(16,activation='relu'),
                                    tf.keras.layers.Dense(1)
                                    ])#model updated to acomedate multiple inputs
```

```python
# Compile the model with an optimizer
sample_model.compile(optimizer='adam',
                     loss=example_loss,
                     metrics=['accuracy'])
```

```python
sample_model.summary()
```

```python
sample_model.fit(X_train,y_train,epochs=10)
```

```python
sample_predict = sample_model.predict(X_test)
```

```python
sample_predict
```

```python
from sklearn.metrics import mean_squared_error 
rmse = (np.sqrt(mean_squared_error(y_test, sample_predict)))
```

```python
rmse
```

```python
mean_squared_error(y_test, sample_predict)
```

```python
# print final accuracy
scores = sample_model.evaluate(X_test, y_test, verbose=0)
print("%s: %.2f%%" % (sample_model.metrics_names[1], scores[1]*100))
```

Maybe inscreasing layers and nodes will improve accuracy?

```python
# Create a Deep Model
sample_model2 = tf.keras.Sequential([
                                    tf.keras.layers.Dense(500,activation='sigmoid',input_shape=[8,]),
                                    tf.keras.layers.Dense(100,activation='relu'),
                                    tf.keras.layers.Dense(100,activation='relu'),
                                    tf.keras.layers.Dense(1)
                                    ])
# Compile the model with an optimizer
sample_model2.compile(optimizer='adam',
                     loss=example_loss,
                     metrics=['accuracy'])
sample_model2.fit(X_train,y_train, epochs=1000)
```

```python
# print final accuracy
scores = sample_model2.evaluate(X_test, y_test, verbose=0)
print("%s: %.2f%%" % (sample_model2.metrics_names[1], scores[1]*100))
```

Accuracy did not change? :(

Maybe less is more? decrease layers and epochs

```python
# Create a Deep Model
sample_model3 = tf.keras.Sequential([
                                    tf.keras.layers.Dense(500,activation='sigmoid',input_shape=[8,]),
                                    tf.keras.layers.Dense(1)
                                    ])
# Compile the model with an optimizer
sample_model3.compile(optimizer='adam',
                     loss=example_loss,
                     metrics=['accuracy'])
sample_model3.fit(X_train,y_train, epochs=100)
```

```python
# print final accuracy
scores = sample_model3.evaluate(X_test, y_test, verbose=0)
print("%s: %.2f%%" % (sample_model3.metrics_names[1], scores[1]*100))
```
